```python
#This cell contains necessary imports
import numpy as np
import pandas as pd
from random import seed
from random import randint
import Crypto
#bytes object initialization 
from Crypto.PublicKey import RSA
#import bcrypt
import rsa
```

```python
#This cell of code spits out ten keys for the agents into a dataframe
cols = ['AgentID', 'PubKey', 'PrivKey']
N = 10
Agent_RSA = pd.DataFrame(columns = cols)
for i in range(N):
    (pubkey, privkey) = rsa.newkeys(4096, poolsize = 6)
    Agent_RSA = Agent_RSA.append({'AgentID' : i, 'PubKey' : pubkey , 'PrivKey' : privkey}, ignore_index = True)
```

```python
Agent_RSA.to_csv('RSA_AgentKeys.csv')
```

```python
def sign_the_block(agentID, block):
    signature = rsa.sign(block.encode('utf-8'), Agent_RSA.loc[agentID].PrivKey, 'SHA-256')
    return signature
```

```python
message = 'This is sparta and now im making this foolishly long'
signed_block = sign_the_block(2, message.encode('utf-8'))
```
